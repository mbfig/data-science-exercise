transform the datetime

```python
training_data['datetime'] = pd.to_datetime(training_data['datetime'])
test_data['datetime'] = pd.to_datetime(test_data['datetime'])
```

```python
training_data['year'] = training_data.datetime.apply(lambda x: x.year)
training_data['month'] = training_data.datetime.apply(lambda x: x.month)
training_data['day'] = training_data.datetime.apply(lambda x: x.day)
training_data['dayofWeek'] = training_data.datetime.apply(lambda x: x.dayofweek)
training_data['weekOfYear'] = training_data.datetime.apply(lambda x: x.weekofyear)
```

```python
test_data['year'] = test_data.datetime.apply(lambda x: x.year)
test_data['month'] = test_data.datetime.apply(lambda x: x.month)
test_data['day'] = test_data.datetime.apply(lambda x: x.day)
test_data['dayofWeek'] = test_data.datetime.apply(lambda x: x.dayofweek)
test_data['WeekOfYear'] = test_data.datetime.apply(lambda x: x.weekofyear)
```

```python
size_week = 2*24*7
```

```python
# have to do shared y axis... they all have different scales

for house_id, pos in zip(training_data.id.unique(), range(0,training_data.id.nunique(), 1)):
    plt.figure(figsize=(20, 30))
    a = plt.subplot(training_data.id.nunique(),1,pos+1)
    aux = training_data[training_data.id == house_id]
    aux.sort_values('datetime')
    plt.plot(aux.datetime[:2*size_week], aux.usage[:2*size_week])
    plt.title('House {}'.format(house_id))
    
    locs, labels = plt.xticks()
    x = list(range(0, len(locs), size_week))
    new_locs = [locs[i] for i in x]
    new_labels = [labels[i] for i in x]
    if pos != training_data.id.nunique()-1:
        plt.setp(a.get_xticklabels(), visible=False)
    plt.xticks(new_locs, rotation='vertical')
    plt.show()
```

#### For most of the houses there is a daily seasonality, and weekends.
#### some houses have periods of low consumption (holidays?)

```python
# given that the test file doesn't have the true value, I'm going to split the training dataset again I'll consider the last 0.33 as test set before I continue
```

```python
n_index = training_data.datetime.nunique()
idx_split = math.floor(n_index*0.66)
list_dates = list(training_data.datetime.unique())
threshold = list_dates[idx_split]
print(list_dates[idx_split])
```

```python
df_training = training_data[training_data.datetime < threshold]
df_test = training_data[training_data.datetime >= threshold]
```

```python
df_training.id.unique()
```

Let's see the trends

```python
for house_id in df_training.id.unique():
#     print('House {}'.format(house_id))
    ts_for_house = df_training[df_training.id == house_id]
    result = seasonal_decompose(
        ts_for_house.usage,
        model='additive',
        freq = 365,
    )
    result.plot()
```

Some houses have a decrese once we enter the 'hot months'. There is a house that start to have heighter consumption on those months

See the autocorrelation values

```python
# have to do shared y axis... they all have different scales

for house_id, pos in zip(training_data.id.unique(), range(0,training_data.id.nunique(), 1)):

    a = plt.subplot(training_data.id.nunique(),1,pos+1)
    aux = training_data[training_data.id == house_id]
    aux.sort_values('datetime')
    plot_acf(aux.usage, lags = 24*2*7)
    plt.title('House {}'.format(house_id))
    plt.show()
    
```

MAC00085 has a lot of correlation within the time series. We can see a daily pattern. It looks that lag-1 is highly correlated with t for all the houses. There is a light daily pattern on most of the houses.
